```python
import pandas as pd
from sklearn.neighbors import KDTree
import numpy as np
```

```python
train_df = pd.read_csv('/content/drive/MyDrive/ML4_adv/최종_위도경도컬럼추가_train.csv')
subway_feature_df = pd.read_csv('/content/drive/MyDrive/ML4_adv/subway_feature.csv')
```

```python
# KDTree에 사용할 좌표를 준비합니다
subway_coords = subway_feature_df[['위도', '경도']].values
subway_tree = KDTree(subway_coords)

# 가장 가까운 지하철역을 찾는 함수 (호선 포함)
def find_nearest_station_kdtree(lat, lon, subway_df, subway_tree):
    distance, index = subway_tree.query([[lat, lon]], k=1)
    nearest_station = subway_df.iloc[index[0]]['역사명']
    nearest_line = subway_df.iloc[index[0]]['호선']
    return nearest_station, nearest_line

# train 데이터에 가장 가까운 지하철역과 호선을 추가합니다
train_df[['nearest_station', 'nearest_line']] = train_df.apply(
    lambda row: pd.Series(find_nearest_station_kdtree(row['Latitude'], row['Longitude'], subway_feature_df, subway_tree)),
    axis=1
)

# 결과를 출력합니다
print(train_df[['시군구', '번지', 'Latitude', 'Longitude', 'nearest_station', 'nearest_line']].head())
```

```python
# 병합된 데이터프레임을 저장합니다
train_df.to_csv('/content/drive/MyDrive/ML4_adv/merged_train_with_nearest_station_and_line_with_distance_columns.csv', index=False)
```
